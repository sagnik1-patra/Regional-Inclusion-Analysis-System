```python
import os
import json
import yaml
import pickle
import warnings
warnings.filterwarnings("ignore")

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.ensemble import RandomForestRegressor

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping

# =========================================================
# PATHS
# =========================================================
csv_path = r"C:\Users\NXTWAVE\Downloads\Regional Inclusion Analysis System\Percentage_of_Scheduled_Caste_Population_to_Total_Population.csv"
output_dir = r"C:\Users\NXTWAVE\Downloads\Regional Inclusion Analysis System"

os.makedirs(output_dir, exist_ok=True)

prefix = "ais_"

# =========================================================
# LOAD DATASET
# =========================================================
df = pd.read_csv(csv_path)

print("Dataset Loaded Successfully")
print(df.head())
print("\nShape:", df.shape)
print("\nColumns:", df.columns.tolist())

# =========================================================
# CLEAN DATA
# =========================================================
df = df.drop_duplicates()

district_col = "District"

for col in df.columns:
    if col != district_col:
        df[col] = pd.to_numeric(df[col], errors="coerce")

df = df.fillna(df.mean(numeric_only=True))

# =========================================================
# COLUMN NAMES
# =========================================================
rural_2001 = "Percentage of Scheduled Caste Population _2001_Rural"
urban_2001 = "Percentage of Scheduled Caste Population _2001_Urban"
total_2001 = "Percentage of Scheduled Caste Population _2001_Total"

rural_2011 = "Percentage of Scheduled Caste Population _2011_Rural"
urban_2011 = "Percentage of Scheduled Caste Population _2011_Urban"
total_2011 = "Percentage of Scheduled Caste Population _2011_Total"

# =========================================================
# FEATURE ENGINEERING
# =========================================================
df["Rural_Growth"] = df[rural_2011] - df[rural_2001]
df["Urban_Growth"] = df[urban_2011] - df[urban_2001]
df["Total_Growth"] = df[total_2011] - df[total_2001]

df["Rural_Urban_Gap_2001"] = df[rural_2001] - df[urban_2001]
df["Rural_Urban_Gap_2011"] = df[rural_2011] - df[urban_2011]

df["Average_SC_Percentage"] = df[
    [rural_2001, urban_2001, total_2001, rural_2011, urban_2011, total_2011]
].mean(axis=1)

df["Priority_Score"] = (
    df[total_2011] * 0.6 +
    df["Total_Growth"] * 0.3 +
    abs(df["Rural_Urban_Gap_2011"]) * 0.1
)

# =========================================================
# AIS PRIORITY CATEGORY
# =========================================================
low_q = df["Priority_Score"].quantile(0.33)
high_q = df["Priority_Score"].quantile(0.66)

def ais_priority_category(score):
    if score <= low_q:
        return "Low Priority"
    elif score <= high_q:
        return "Medium Priority"
    else:
        return "High Priority"

df["AIS_Priority_Category"] = df["Priority_Score"].apply(ais_priority_category)

# =========================================================
# FIXED AIS OPTIMIZATION FUNCTION
# =========================================================
def ais_feature_weight_optimization(
    X_train,
    X_test,
    y_train,
    y_test,
    population_size=30,
    selected_size=5,
    clone_factor=4,
    mutation_rate=0.15,
    iterations=30
):
    n_features = X_train.shape[1]

    population = np.random.uniform(
        0.5,
        1.5,
        size=(population_size, n_features)
    )

    best_antibody = None
    best_score = -999999
    convergence = []

    for iteration in range(iterations):
        scores = []

        for antibody in population:
            X_train_weighted = X_train * antibody
            X_test_weighted = X_test * antibody

            model = RandomForestRegressor(
                n_estimators=100,
                random_state=42
            )

            model.fit(X_train_weighted, y_train)
            preds = model.predict(X_test_weighted)

            r2 = r2_score(y_test, preds)
            scores.append(r2)

        scores = np.array(scores)

        best_index = np.argmax(scores)

        if scores[best_index] > best_score:
            best_score = scores[best_index]
            best_antibody = population[best_index].copy()

        convergence.append(best_score)

        selected_indices = scores.argsort()[-selected_size:]
        selected = population[selected_indices]

        clones = []
        max_clones = int(population_size * 0.7)

        for antibody in selected:
            for _ in range(clone_factor):
                if len(clones) >= max_clones:
                    break

                clone = antibody.copy()
                mutation = np.random.normal(
                    0,
                    mutation_rate,
                    size=n_features
                )

                clone = clone + mutation
                clone = np.clip(clone, 0.1, 2.0)
                clones.append(clone)

        clones = np.array(clones)

        remaining_count = population_size - len(clones)

        if remaining_count > 0:
            random_new = np.random.uniform(
                0.5,
                1.5,
                size=(remaining_count, n_features)
            )

            population = np.vstack([clones, random_new])
        else:
            population = clones[:population_size]

    return best_antibody, best_score, convergence

# =========================================================
# ENCODE DISTRICT
# =========================================================
district_encoder = LabelEncoder()
df["District_Encoded"] = district_encoder.fit_transform(df[district_col])

# =========================================================
# FEATURES AND TARGET
# =========================================================
target_col = total_2011

feature_cols = [
    "District_Encoded",
    rural_2001,
    urban_2001,
    total_2001,
    rural_2011,
    urban_2011,
    "Rural_Growth",
    "Urban_Growth",
    "Total_Growth",
    "Rural_Urban_Gap_2001",
    "Rural_Urban_Gap_2011",
    "Average_SC_Percentage",
    "Priority_Score"
]

X = df[feature_cols]
y = df[target_col]

# =========================================================
# TRAIN TEST SPLIT
# =========================================================
X_train, X_test, y_train, y_test, district_train, district_test = train_test_split(
    X,
    y,
    df[district_col],
    test_size=0.2,
    random_state=42
)

# =========================================================
# SCALING
# =========================================================
scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# =========================================================
# AIS FEATURE WEIGHT OPTIMIZATION
# =========================================================
best_weights, best_ais_score, convergence = ais_feature_weight_optimization(
    X_train_scaled,
    X_test_scaled,
    y_train,
    y_test
)

X_train_ais = X_train_scaled * best_weights
X_test_ais = X_test_scaled * best_weights

# =========================================================
# AIS RANDOM FOREST MODEL
# =========================================================
ais_rf_model = RandomForestRegressor(
    n_estimators=300,
    random_state=42
)

ais_rf_model.fit(X_train_ais, y_train)

ais_rf_predictions = ais_rf_model.predict(X_test_ais)

ais_rf_mae = mean_absolute_error(y_test, ais_rf_predictions)
ais_rf_mse = mean_squared_error(y_test, ais_rf_predictions)
ais_rf_rmse = np.sqrt(ais_rf_mse)
ais_rf_r2 = r2_score(y_test, ais_rf_predictions)
ais_rf_accuracy = max(0, ais_rf_r2 * 100)

# =========================================================
# AIS DEEP LEARNING MODEL
# =========================================================
ais_dl_model = Sequential([
    Dense(64, activation="relu", input_shape=(X_train_ais.shape[1],)),
    Dropout(0.2),
    Dense(32, activation="relu"),
    Dense(16, activation="relu"),
    Dense(1)
])

ais_dl_model.compile(
    optimizer="adam",
    loss="mse",
    metrics=["mae"]
)

early_stop = EarlyStopping(
    monitor="val_loss",
    patience=20,
    restore_best_weights=True
)

history = ais_dl_model.fit(
    X_train_ais,
    y_train,
    validation_split=0.2,
    epochs=300,
    batch_size=8,
    callbacks=[early_stop],
    verbose=1
)

ais_dl_predictions = ais_dl_model.predict(X_test_ais).flatten()

ais_dl_mae = mean_absolute_error(y_test, ais_dl_predictions)
ais_dl_mse = mean_squared_error(y_test, ais_dl_predictions)
ais_dl_rmse = np.sqrt(ais_dl_mse)
ais_dl_r2 = r2_score(y_test, ais_dl_predictions)
ais_dl_accuracy = max(0, ais_dl_r2 * 100)

# =========================================================
# RESULTS CSV
# =========================================================
results_df = pd.DataFrame([
    {
        "Model": "AIS Random Forest",
        "MAE": ais_rf_mae,
        "MSE": ais_rf_mse,
        "RMSE": ais_rf_rmse,
        "R2_Score": ais_rf_r2,
        "Accuracy_Percentage": ais_rf_accuracy
    },
    {
        "Model": "AIS Deep Learning",
        "MAE": ais_dl_mae,
        "MSE": ais_dl_mse,
        "RMSE": ais_dl_rmse,
        "R2_Score": ais_dl_r2,
        "Accuracy_Percentage": ais_dl_accuracy
    }
])

result_csv_path = os.path.join(output_dir, prefix + "result.csv")
results_df.to_csv(result_csv_path, index=False)

print("\nAIS Model Results")
print(results_df)

# =========================================================
# PREDICTION CSV
# =========================================================
prediction_df = pd.DataFrame({
    "District": district_test.values,
    "Actual_2011_Total_SC_Percentage": y_test.values,
    "AIS_RF_Prediction": ais_rf_predictions,
    "AIS_DL_Prediction": ais_dl_predictions
})

prediction_csv_path = os.path.join(output_dir, prefix + "prediction.csv")
prediction_df.to_csv(prediction_csv_path, index=False)

print("\nAIS Prediction Results")
print(prediction_df.head())

# =========================================================
# FULL DATASET RESULT CSV
# =========================================================
full_scaled = scaler.transform(X)
full_ais = full_scaled * best_weights

df["AIS_RF_Full_Prediction"] = ais_rf_model.predict(full_ais)
df["AIS_Prediction_Error"] = abs(df[target_col] - df["AIS_RF_Full_Prediction"])

full_result_csv_path = os.path.join(output_dir, prefix + "full_dataset_result.csv")
df.to_csv(full_result_csv_path, index=False)

# =========================================================
# SAVE H5 MODEL
# =========================================================
h5_path = os.path.join(output_dir, prefix + "model.h5")
ais_dl_model.save(h5_path)

# =========================================================
# SAVE PKL MODEL
# =========================================================
pkl_path = os.path.join(output_dir, prefix + "model.pkl")

pkl_data = {
    "ais_random_forest_model": ais_rf_model,
    "scaler": scaler,
    "district_encoder": district_encoder,
    "best_ais_feature_weights": best_weights,
    "feature_columns": feature_cols,
    "target_column": target_col,
    "ais_priority_categories": {
        "low_quantile": float(low_q),
        "high_quantile": float(high_q)
    }
}

with open(pkl_path, "wb") as f:
    pickle.dump(pkl_data, f)

# =========================================================
# SAVE JSON
# =========================================================
json_path = os.path.join(output_dir, prefix + "results.json")

json_data = {
    "project_name": "AIS Smart Regional Inclusion Analysis System",
    "author": "Sagnik Patra",
    "algorithm": "Artificial Immune System",
    "dataset_path": csv_path,
    "output_directory": output_dir,
    "target_column": target_col,
    "feature_columns": feature_cols,
    "ais_best_score": float(best_ais_score),
    "ais_best_feature_weights": best_weights.tolist(),
    "ais_convergence": [float(x) for x in convergence],
    "model_results": results_df.to_dict(orient="records"),
    "files_saved": {
        "h5_model": h5_path,
        "pkl_model": pkl_path,
        "result_csv": result_csv_path,
        "prediction_csv": prediction_csv_path,
        "full_dataset_result_csv": full_result_csv_path
    }
}

with open(json_path, "w") as f:
    json.dump(json_data, f, indent=4)

# =========================================================
# SAVE YAML
# =========================================================
yaml_path = os.path.join(output_dir, prefix + "config.yaml")

yaml_data = {
    "project": "AIS Smart Regional Inclusion Analysis System",
    "author": "Sagnik Patra",
    "algorithm": "Artificial Immune System",
    "model_type": "Regression",
    "target": target_col,
    "features": feature_cols,
    "ais_best_score": float(best_ais_score),
    "outputs": {
        "h5": h5_path,
        "pkl": pkl_path,
        "json": json_path,
        "yaml": yaml_path,
        "result_csv": result_csv_path,
        "prediction_csv": prediction_csv_path,
        "full_dataset_result_csv": full_result_csv_path
    }
}

with open(yaml_path, "w") as f:
    yaml.dump(yaml_data, f, default_flow_style=False)

# =========================================================
# GRAPH 1: AIS ACCURACY GRAPH
# =========================================================
plt.figure(figsize=(10, 6))
plt.bar(results_df["Model"], results_df["Accuracy_Percentage"])
plt.title("AIS Model Accuracy Graph")
plt.xlabel("Model")
plt.ylabel("Accuracy Percentage")
plt.xticks(rotation=20)
plt.tight_layout()

accuracy_graph_path = os.path.join(output_dir, prefix + "accuracy_graph.png")
plt.savefig(accuracy_graph_path, dpi=300)
plt.show()

# =========================================================
# GRAPH 2: AIS HEATMAP
# =========================================================
corr = df[feature_cols + [target_col]].corr()

plt.figure(figsize=(12, 8))
plt.imshow(corr, cmap="coolwarm", aspect="auto")
plt.colorbar()
plt.xticks(range(len(corr.columns)), corr.columns, rotation=90)
plt.yticks(range(len(corr.columns)), corr.columns)
plt.title("AIS Feature Correlation Heatmap")
plt.tight_layout()

heatmap_path = os.path.join(output_dir, prefix + "heatmap.png")
plt.savefig(heatmap_path, dpi=300)
plt.show()

# =========================================================
# GRAPH 3: AIS COMPARISON GRAPH
# =========================================================
plt.figure(figsize=(10, 6))
plt.plot(results_df["Model"], results_df["MAE"], marker="o", label="MAE")
plt.plot(results_df["Model"], results_df["RMSE"], marker="o", label="RMSE")
plt.title("AIS Model Error Comparison Graph")
plt.xlabel("Model")
plt.ylabel("Error Value")
plt.xticks(rotation=20)
plt.legend()
plt.tight_layout()

comparison_graph_path = os.path.join(output_dir, prefix + "comparison_graph.png")
plt.savefig(comparison_graph_path, dpi=300)
plt.show()

# =========================================================
# GRAPH 4: AIS RESULT GRAPH
# =========================================================
plt.figure(figsize=(10, 6))
plt.bar(results_df["Model"], results_df["R2_Score"])
plt.title("AIS R2 Score Result Graph")
plt.xlabel("Model")
plt.ylabel("R2 Score")
plt.xticks(rotation=20)
plt.tight_layout()

result_graph_path = os.path.join(output_dir, prefix + "result_graph.png")
plt.savefig(result_graph_path, dpi=300)
plt.show()

# =========================================================
# GRAPH 5: AIS PREDICTION GRAPH
# =========================================================
plt.figure(figsize=(12, 6))
plt.plot(prediction_df["Actual_2011_Total_SC_Percentage"].values, marker="o", label="Actual")
plt.plot(prediction_df["AIS_RF_Prediction"].values, marker="o", label="AIS RF Predicted")
plt.title("AIS Actual vs Predicted Graph")
plt.xlabel("Test Samples")
plt.ylabel("2011 Total SC Percentage")
plt.legend()
plt.tight_layout()

prediction_graph_path = os.path.join(output_dir, prefix + "prediction_graph.png")
plt.savefig(prediction_graph_path, dpi=300)
plt.show()

# =========================================================
# GRAPH 6: AIS CONVERGENCE GRAPH
# =========================================================
plt.figure(figsize=(10, 6))
plt.plot(convergence, marker="o")
plt.title("AIS Convergence Graph")
plt.xlabel("Iteration")
plt.ylabel("Best R2 Score")
plt.tight_layout()

convergence_graph_path = os.path.join(output_dir, prefix + "convergence_graph.png")
plt.savefig(convergence_graph_path, dpi=300)
plt.show()

# =========================================================
# GRAPH 7: AIS GROWTH GRAPH
# =========================================================
top_growth_df = df.sort_values(by="Total_Growth", ascending=False).head(15)

plt.figure(figsize=(12, 6))
plt.bar(top_growth_df[district_col], top_growth_df["Total_Growth"])
plt.title("AIS Top 15 Districts by SC Population Growth")
plt.xlabel("District")
plt.ylabel("Total Growth")
plt.xticks(rotation=75)
plt.tight_layout()

growth_graph_path = os.path.join(output_dir, prefix + "growth_graph.png")
plt.savefig(growth_graph_path, dpi=300)
plt.show()

# =========================================================
# GRAPH 8: AIS TRAINING GRAPH
# =========================================================
plt.figure(figsize=(10, 6))
plt.plot(history.history["loss"], label="Training Loss")
plt.plot(history.history["val_loss"], label="Validation Loss")
plt.title("AIS Deep Learning Training Loss Graph")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.legend()
plt.tight_layout()

training_graph_path = os.path.join(output_dir, prefix + "training_graph.png")
plt.savefig(training_graph_path, dpi=300)
plt.show()

# =========================================================
# FINAL OUTPUT
# =========================================================
print("\nAll AIS files saved successfully")
print("H5 Model:", h5_path)
print("PKL Model:", pkl_path)
print("JSON File:", json_path)
print("YAML File:", yaml_path)
print("Result CSV:", result_csv_path)
print("Prediction CSV:", prediction_csv_path)
print("Full Dataset Result CSV:", full_result_csv_path)
print("Accuracy Graph:", accuracy_graph_path)
print("Heatmap:", heatmap_path)
print("Comparison Graph:", comparison_graph_path)
print("Result Graph:", result_graph_path)
print("Prediction Graph:", prediction_graph_path)
print("Convergence Graph:", convergence_graph_path)
print("Growth Graph:", growth_graph_path)
print("Training Graph:", training_graph_path)
```
