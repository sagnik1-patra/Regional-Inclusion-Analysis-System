```python
import os
import json
import yaml
import pickle
import warnings
warnings.filterwarnings("ignore")

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.ensemble import RandomForestRegressor

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping

# =========================================================
# PATHS
# =========================================================
csv_path = r"C:\Users\NXTWAVE\Downloads\Regional Inclusion Analysis System\Percentage_of_Scheduled_Caste_Population_to_Total_Population.csv"
output_dir = r"C:\Users\NXTWAVE\Downloads\Regional Inclusion Analysis System"

os.makedirs(output_dir, exist_ok=True)

prefix = "pso_"

# =========================================================
# LOAD DATASET
# =========================================================
df = pd.read_csv(csv_path)

print("Dataset Loaded Successfully")
print(df.head())
print("\nShape:", df.shape)
print("\nColumns:", df.columns.tolist())

# =========================================================
# CLEAN DATA
# =========================================================
df = df.drop_duplicates()

district_col = "District"

for col in df.columns:
    if col != district_col:
        df[col] = pd.to_numeric(df[col], errors="coerce")

df = df.fillna(df.mean(numeric_only=True))

# =========================================================
# COLUMN NAMES
# =========================================================
rural_2001 = "Percentage of Scheduled Caste Population _2001_Rural"
urban_2001 = "Percentage of Scheduled Caste Population _2001_Urban"
total_2001 = "Percentage of Scheduled Caste Population _2001_Total"

rural_2011 = "Percentage of Scheduled Caste Population _2011_Rural"
urban_2011 = "Percentage of Scheduled Caste Population _2011_Urban"
total_2011 = "Percentage of Scheduled Caste Population _2011_Total"

# =========================================================
# FEATURE ENGINEERING
# =========================================================
df["Rural_Growth"] = df[rural_2011] - df[rural_2001]
df["Urban_Growth"] = df[urban_2011] - df[urban_2001]
df["Total_Growth"] = df[total_2011] - df[total_2001]

df["Rural_Urban_Gap_2001"] = df[rural_2001] - df[urban_2001]
df["Rural_Urban_Gap_2011"] = df[rural_2011] - df[urban_2011]

df["Average_SC_Percentage"] = df[
    [rural_2001, urban_2001, total_2001, rural_2011, urban_2011, total_2011]
].mean(axis=1)

df["Priority_Score"] = (
    df[total_2011] * 0.6 +
    df["Total_Growth"] * 0.3 +
    abs(df["Rural_Urban_Gap_2011"]) * 0.1
)

# =========================================================
# PSO PRIORITY CATEGORY
# =========================================================
low_q = df["Priority_Score"].quantile(0.33)
high_q = df["Priority_Score"].quantile(0.66)

def pso_priority_category(score):
    if score <= low_q:
        return "Low Priority"
    elif score <= high_q:
        return "Medium Priority"
    else:
        return "High Priority"

df["PSO_Priority_Category"] = df["Priority_Score"].apply(pso_priority_category)

# =========================================================
# PSO OPTIMIZATION FUNCTION
# =========================================================
def pso_feature_weight_optimization(
    X_train,
    X_test,
    y_train,
    y_test,
    num_particles=30,
    iterations=30,
    inertia_weight=0.7,
    cognitive_weight=1.5,
    social_weight=1.5
):
    """
    Particle Swarm Optimization for feature-weight selection.
    Each particle represents one feature-weight vector.
    Fitness = R2 score of Random Forest model.
    """

    n_features = X_train.shape[1]

    positions = np.random.uniform(
        0.5,
        1.5,
        size=(num_particles, n_features)
    )

    velocities = np.random.uniform(
        -0.1,
        0.1,
        size=(num_particles, n_features)
    )

    personal_best_positions = positions.copy()
    personal_best_scores = np.full(num_particles, -999999.0)

    global_best_position = None
    global_best_score = -999999.0

    convergence = []

    for iteration in range(iterations):
        for i in range(num_particles):
            particle_weights = positions[i]

            X_train_weighted = X_train * particle_weights
            X_test_weighted = X_test * particle_weights

            model = RandomForestRegressor(
                n_estimators=100,
                random_state=42
            )

            model.fit(X_train_weighted, y_train)
            preds = model.predict(X_test_weighted)

            score = r2_score(y_test, preds)

            if score > personal_best_scores[i]:
                personal_best_scores[i] = score
                personal_best_positions[i] = positions[i].copy()

            if score > global_best_score:
                global_best_score = score
                global_best_position = positions[i].copy()

        convergence.append(global_best_score)

        for i in range(num_particles):
            r1 = np.random.rand(n_features)
            r2 = np.random.rand(n_features)

            cognitive_velocity = cognitive_weight * r1 * (
                personal_best_positions[i] - positions[i]
            )

            social_velocity = social_weight * r2 * (
                global_best_position - positions[i]
            )

            velocities[i] = (
                inertia_weight * velocities[i]
                + cognitive_velocity
                + social_velocity
            )

            positions[i] = positions[i] + velocities[i]
            positions[i] = np.clip(positions[i], 0.1, 2.0)

    return global_best_position, global_best_score, convergence

# =========================================================
# ENCODE DISTRICT
# =========================================================
district_encoder = LabelEncoder()
df["District_Encoded"] = district_encoder.fit_transform(df[district_col])

# =========================================================
# FEATURES AND TARGET
# =========================================================
target_col = total_2011

feature_cols = [
    "District_Encoded",
    rural_2001,
    urban_2001,
    total_2001,
    rural_2011,
    urban_2011,
    "Rural_Growth",
    "Urban_Growth",
    "Total_Growth",
    "Rural_Urban_Gap_2001",
    "Rural_Urban_Gap_2011",
    "Average_SC_Percentage",
    "Priority_Score"
]

X = df[feature_cols]
y = df[target_col]

# =========================================================
# TRAIN TEST SPLIT
# =========================================================
X_train, X_test, y_train, y_test, district_train, district_test = train_test_split(
    X,
    y,
    df[district_col],
    test_size=0.2,
    random_state=42
)

# =========================================================
# SCALING
# =========================================================
scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# =========================================================
# PSO FEATURE WEIGHT OPTIMIZATION
# =========================================================
best_weights, best_pso_score, convergence = pso_feature_weight_optimization(
    X_train_scaled,
    X_test_scaled,
    y_train,
    y_test
)

X_train_pso = X_train_scaled * best_weights
X_test_pso = X_test_scaled * best_weights

# =========================================================
# PSO RANDOM FOREST MODEL
# =========================================================
pso_rf_model = RandomForestRegressor(
    n_estimators=300,
    random_state=42
)

pso_rf_model.fit(X_train_pso, y_train)

pso_rf_predictions = pso_rf_model.predict(X_test_pso)

pso_rf_mae = mean_absolute_error(y_test, pso_rf_predictions)
pso_rf_mse = mean_squared_error(y_test, pso_rf_predictions)
pso_rf_rmse = np.sqrt(pso_rf_mse)
pso_rf_r2 = r2_score(y_test, pso_rf_predictions)
pso_rf_accuracy = max(0, pso_rf_r2 * 100)

# =========================================================
# PSO DEEP LEARNING MODEL
# =========================================================
pso_dl_model = Sequential([
    Dense(64, activation="relu", input_shape=(X_train_pso.shape[1],)),
    Dropout(0.2),
    Dense(32, activation="relu"),
    Dense(16, activation="relu"),
    Dense(1)
])

pso_dl_model.compile(
    optimizer="adam",
    loss="mse",
    metrics=["mae"]
)

early_stop = EarlyStopping(
    monitor="val_loss",
    patience=20,
    restore_best_weights=True
)

history = pso_dl_model.fit(
    X_train_pso,
    y_train,
    validation_split=0.2,
    epochs=300,
    batch_size=8,
    callbacks=[early_stop],
    verbose=1
)

pso_dl_predictions = pso_dl_model.predict(X_test_pso).flatten()

pso_dl_mae = mean_absolute_error(y_test, pso_dl_predictions)
pso_dl_mse = mean_squared_error(y_test, pso_dl_predictions)
pso_dl_rmse = np.sqrt(pso_dl_mse)
pso_dl_r2 = r2_score(y_test, pso_dl_predictions)
pso_dl_accuracy = max(0, pso_dl_r2 * 100)

# =========================================================
# RESULT CSV
# =========================================================
results_df = pd.DataFrame([
    {
        "Model": "PSO Random Forest",
        "MAE": pso_rf_mae,
        "MSE": pso_rf_mse,
        "RMSE": pso_rf_rmse,
        "R2_Score": pso_rf_r2,
        "Accuracy_Percentage": pso_rf_accuracy
    },
    {
        "Model": "PSO Deep Learning",
        "MAE": pso_dl_mae,
        "MSE": pso_dl_mse,
        "RMSE": pso_dl_rmse,
        "R2_Score": pso_dl_r2,
        "Accuracy_Percentage": pso_dl_accuracy
    }
])

result_csv_path = os.path.join(output_dir, prefix + "result.csv")
results_df.to_csv(result_csv_path, index=False)

print("\nPSO Model Results")
print(results_df)

# =========================================================
# PREDICTION CSV
# =========================================================
prediction_df = pd.DataFrame({
    "District": district_test.values,
    "Actual_2011_Total_SC_Percentage": y_test.values,
    "PSO_RF_Prediction": pso_rf_predictions,
    "PSO_DL_Prediction": pso_dl_predictions
})

prediction_csv_path = os.path.join(output_dir, prefix + "prediction.csv")
prediction_df.to_csv(prediction_csv_path, index=False)

print("\nPSO Prediction Results")
print(prediction_df.head())

# =========================================================
# FULL DATASET RESULT CSV
# =========================================================
full_scaled = scaler.transform(X)
full_pso = full_scaled * best_weights

df["PSO_RF_Full_Prediction"] = pso_rf_model.predict(full_pso)
df["PSO_Prediction_Error"] = abs(df[target_col] - df["PSO_RF_Full_Prediction"])

full_result_csv_path = os.path.join(output_dir, prefix + "full_dataset_result.csv")
df.to_csv(full_result_csv_path, index=False)

# =========================================================
# SAVE H5 MODEL
# =========================================================
h5_path = os.path.join(output_dir, prefix + "model.h5")
pso_dl_model.save(h5_path)

# =========================================================
# SAVE PKL MODEL
# =========================================================
pkl_path = os.path.join(output_dir, prefix + "model.pkl")

pkl_data = {
    "pso_random_forest_model": pso_rf_model,
    "scaler": scaler,
    "district_encoder": district_encoder,
    "best_pso_feature_weights": best_weights,
    "feature_columns": feature_cols,
    "target_column": target_col,
    "pso_priority_categories": {
        "low_quantile": float(low_q),
        "high_quantile": float(high_q)
    }
}

with open(pkl_path, "wb") as f:
    pickle.dump(pkl_data, f)

# =========================================================
# SAVE JSON
# =========================================================
json_path = os.path.join(output_dir, prefix + "results.json")

json_data = {
    "project_name": "PSO Smart Regional Inclusion Analysis System",
    "author": "Sagnik Patra",
    "algorithm": "Particle Swarm Optimization",
    "dataset_path": csv_path,
    "output_directory": output_dir,
    "target_column": target_col,
    "feature_columns": feature_cols,
    "pso_best_score": float(best_pso_score),
    "pso_best_feature_weights": best_weights.tolist(),
    "pso_convergence": [float(x) for x in convergence],
    "model_results": results_df.to_dict(orient="records"),
    "files_saved": {
        "h5_model": h5_path,
        "pkl_model": pkl_path,
        "result_csv": result_csv_path,
        "prediction_csv": prediction_csv_path,
        "full_dataset_result_csv": full_result_csv_path
    }
}

with open(json_path, "w") as f:
    json.dump(json_data, f, indent=4)

# =========================================================
# SAVE YAML
# =========================================================
yaml_path = os.path.join(output_dir, prefix + "config.yaml")

yaml_data = {
    "project": "PSO Smart Regional Inclusion Analysis System",
    "author": "Sagnik Patra",
    "algorithm": "Particle Swarm Optimization",
    "model_type": "Regression",
    "target": target_col,
    "features": feature_cols,
    "pso_best_score": float(best_pso_score),
    "outputs": {
        "h5": h5_path,
        "pkl": pkl_path,
        "json": json_path,
        "yaml": yaml_path,
        "result_csv": result_csv_path,
        "prediction_csv": prediction_csv_path,
        "full_dataset_result_csv": full_result_csv_path
    }
}

with open(yaml_path, "w") as f:
    yaml.dump(yaml_data, f, default_flow_style=False)

# =========================================================
# GRAPH 1: PSO ACCURACY GRAPH
# =========================================================
plt.figure(figsize=(10, 6))
plt.bar(results_df["Model"], results_df["Accuracy_Percentage"])
plt.title("PSO Model Accuracy Graph")
plt.xlabel("Model")
plt.ylabel("Accuracy Percentage")
plt.xticks(rotation=20)
plt.tight_layout()

accuracy_graph_path = os.path.join(output_dir, prefix + "accuracy_graph.png")
plt.savefig(accuracy_graph_path, dpi=300)
plt.show()

# =========================================================
# GRAPH 2: PSO HEATMAP
# =========================================================
corr = df[feature_cols + [target_col]].corr()

plt.figure(figsize=(12, 8))
plt.imshow(corr, cmap="coolwarm", aspect="auto")
plt.colorbar()
plt.xticks(range(len(corr.columns)), corr.columns, rotation=90)
plt.yticks(range(len(corr.columns)), corr.columns)
plt.title("PSO Feature Correlation Heatmap")
plt.tight_layout()

heatmap_path = os.path.join(output_dir, prefix + "heatmap.png")
plt.savefig(heatmap_path, dpi=300)
plt.show()

# =========================================================
# GRAPH 3: PSO COMPARISON GRAPH
# =========================================================
plt.figure(figsize=(10, 6))
plt.plot(results_df["Model"], results_df["MAE"], marker="o", label="MAE")
plt.plot(results_df["Model"], results_df["RMSE"], marker="o", label="RMSE")
plt.title("PSO Model Error Comparison Graph")
plt.xlabel("Model")
plt.ylabel("Error Value")
plt.xticks(rotation=20)
plt.legend()
plt.tight_layout()

comparison_graph_path = os.path.join(output_dir, prefix + "comparison_graph.png")
plt.savefig(comparison_graph_path, dpi=300)
plt.show()

# =========================================================
# GRAPH 4: PSO RESULT GRAPH
# =========================================================
plt.figure(figsize=(10, 6))
plt.bar(results_df["Model"], results_df["R2_Score"])
plt.title("PSO R2 Score Result Graph")
plt.xlabel("Model")
plt.ylabel("R2 Score")
plt.xticks(rotation=20)
plt.tight_layout()

result_graph_path = os.path.join(output_dir, prefix + "result_graph.png")
plt.savefig(result_graph_path, dpi=300)
plt.show()

# =========================================================
# GRAPH 5: PSO PREDICTION GRAPH
# =========================================================
plt.figure(figsize=(12, 6))
plt.plot(prediction_df["Actual_2011_Total_SC_Percentage"].values, marker="o", label="Actual")
plt.plot(prediction_df["PSO_RF_Prediction"].values, marker="o", label="PSO RF Predicted")
plt.title("PSO Actual vs Predicted Graph")
plt.xlabel("Test Samples")
plt.ylabel("2011 Total SC Percentage")
plt.legend()
plt.tight_layout()

prediction_graph_path = os.path.join(output_dir, prefix + "prediction_graph.png")
plt.savefig(prediction_graph_path, dpi=300)
plt.show()

# =========================================================
# GRAPH 6: PSO CONVERGENCE GRAPH
# =========================================================
plt.figure(figsize=(10, 6))
plt.plot(convergence, marker="o")
plt.title("PSO Convergence Graph")
plt.xlabel("Iteration")
plt.ylabel("Best R2 Score")
plt.tight_layout()

convergence_graph_path = os.path.join(output_dir, prefix + "convergence_graph.png")
plt.savefig(convergence_graph_path, dpi=300)
plt.show()

# =========================================================
# GRAPH 7: PSO GROWTH GRAPH
# =========================================================
top_growth_df = df.sort_values(by="Total_Growth", ascending=False).head(15)

plt.figure(figsize=(12, 6))
plt.bar(top_growth_df[district_col], top_growth_df["Total_Growth"])
plt.title("PSO Top 15 Districts by SC Population Growth")
plt.xlabel("District")
plt.ylabel("Total Growth")
plt.xticks(rotation=75)
plt.tight_layout()

growth_graph_path = os.path.join(output_dir, prefix + "growth_graph.png")
plt.savefig(growth_graph_path, dpi=300)
plt.show()

# =========================================================
# GRAPH 8: PSO TRAINING GRAPH
# =========================================================
plt.figure(figsize=(10, 6))
plt.plot(history.history["loss"], label="Training Loss")
plt.plot(history.history["val_loss"], label="Validation Loss")
plt.title("PSO Deep Learning Training Loss Graph")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.legend()
plt.tight_layout()

training_graph_path = os.path.join(output_dir, prefix + "training_graph.png")
plt.savefig(training_graph_path, dpi=300)
plt.show()

# =========================================================
# FINAL OUTPUT
# =========================================================
print("\nAll PSO files saved successfully")
print("H5 Model:", h5_path)
print("PKL Model:", pkl_path)
print("JSON File:", json_path)
print("YAML File:", yaml_path)
print("Result CSV:", result_csv_path)
print("Prediction CSV:", prediction_csv_path)
print("Full Dataset Result CSV:", full_result_csv_path)
print("Accuracy Graph:", accuracy_graph_path)
print("Heatmap:", heatmap_path)
print("Comparison Graph:", comparison_graph_path)
print("Result Graph:", result_graph_path)
print("Prediction Graph:", prediction_graph_path)
print("Convergence Graph:", convergence_graph_path)
print("Growth Graph:", growth_graph_path)
print("Training Graph:", training_graph_path)
```
