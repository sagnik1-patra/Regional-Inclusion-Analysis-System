```python
import os
import json
import yaml
import pickle
import warnings
warnings.filterwarnings("ignore")

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping

# =========================================================
# PATHS
# =========================================================
csv_path = r"C:\Users\NXTWAVE\Downloads\Regional Inclusion Analysis System\Percentage_of_Scheduled_Caste_Population_to_Total_Population.csv"
output_dir = r"C:\Users\NXTWAVE\Downloads\Regional Inclusion Analysis System"

os.makedirs(output_dir, exist_ok=True)

# =========================================================
# LOAD DATASET
# =========================================================
df = pd.read_csv(csv_path)

print("Dataset Loaded Successfully")
print(df.head())
print("\nDataset Shape:", df.shape)
print("\nColumns:", df.columns.tolist())

# =========================================================
# CLEAN DATA
# =========================================================
df = df.drop_duplicates()

district_col = "District"

for col in df.columns:
    if col != district_col:
        df[col] = pd.to_numeric(df[col], errors="coerce")

df = df.fillna(df.mean(numeric_only=True))

# =========================================================
# COLUMN NAMES
# =========================================================
rural_2001 = "Percentage of Scheduled Caste Population _2001_Rural"
urban_2001 = "Percentage of Scheduled Caste Population _2001_Urban"
total_2001 = "Percentage of Scheduled Caste Population _2001_Total"

rural_2011 = "Percentage of Scheduled Caste Population _2011_Rural"
urban_2011 = "Percentage of Scheduled Caste Population _2011_Urban"
total_2011 = "Percentage of Scheduled Caste Population _2011_Total"

# =========================================================
# FEATURE ENGINEERING
# =========================================================
df["Rural_Growth"] = df[rural_2011] - df[rural_2001]
df["Urban_Growth"] = df[urban_2011] - df[urban_2001]
df["Total_Growth"] = df[total_2011] - df[total_2001]

df["Rural_Urban_Gap_2001"] = df[rural_2001] - df[urban_2001]
df["Rural_Urban_Gap_2011"] = df[rural_2011] - df[urban_2011]

df["Average_SC_Percentage"] = df[
    [rural_2001, urban_2001, total_2001, rural_2011, urban_2011, total_2011]
].mean(axis=1)

df["Priority_Score"] = (
    df[total_2011] * 0.6 +
    df["Total_Growth"] * 0.3 +
    abs(df["Rural_Urban_Gap_2011"]) * 0.1
)

# =========================================================
# ENCODE DISTRICT
# =========================================================
district_encoder = LabelEncoder()
df["District_Encoded"] = district_encoder.fit_transform(df[district_col])

# =========================================================
# FEATURES AND TARGET
# =========================================================
target_col = total_2011

feature_cols = [
    "District_Encoded",
    rural_2001,
    urban_2001,
    total_2001,
    rural_2011,
    urban_2011,
    "Rural_Growth",
    "Urban_Growth",
    "Total_Growth",
    "Rural_Urban_Gap_2001",
    "Rural_Urban_Gap_2011",
    "Average_SC_Percentage",
    "Priority_Score"
]

X = df[feature_cols]
y = df[target_col]

# =========================================================
# TRAIN TEST SPLIT
# =========================================================
X_train, X_test, y_train, y_test, district_train, district_test = train_test_split(
    X,
    y,
    df[district_col],
    test_size=0.2,
    random_state=42
)

# =========================================================
# SCALING
# =========================================================
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# =========================================================
# MODEL TRAINING
# =========================================================
models = {
    "Linear Regression": LinearRegression(),
    "Random Forest": RandomForestRegressor(n_estimators=200, random_state=42),
    "Gradient Boosting": GradientBoostingRegressor(random_state=42)
}

results = []
predictions_dict = {}

for name, model in models.items():
    model.fit(X_train_scaled, y_train)
    preds = model.predict(X_test_scaled)

    mae = mean_absolute_error(y_test, preds)
    mse = mean_squared_error(y_test, preds)
    rmse = np.sqrt(mse)
    r2 = r2_score(y_test, preds)

    accuracy_percentage = max(0, r2 * 100)

    results.append({
        "Model": name,
        "MAE": mae,
        "MSE": mse,
        "RMSE": rmse,
        "R2_Score": r2,
        "Accuracy_Percentage": accuracy_percentage
    })

    predictions_dict[name] = preds

# =========================================================
# DEEP LEARNING MODEL
# =========================================================
dl_model = Sequential([
    Dense(64, activation="relu", input_shape=(X_train_scaled.shape[1],)),
    Dropout(0.2),
    Dense(32, activation="relu"),
    Dense(16, activation="relu"),
    Dense(1)
])

dl_model.compile(
    optimizer="adam",
    loss="mse",
    metrics=["mae"]
)

early_stop = EarlyStopping(
    monitor="val_loss",
    patience=20,
    restore_best_weights=True
)

history = dl_model.fit(
    X_train_scaled,
    y_train,
    validation_split=0.2,
    epochs=300,
    batch_size=8,
    callbacks=[early_stop],
    verbose=1
)

dl_preds = dl_model.predict(X_test_scaled).flatten()

dl_mae = mean_absolute_error(y_test, dl_preds)
dl_mse = mean_squared_error(y_test, dl_preds)
dl_rmse = np.sqrt(dl_mse)
dl_r2 = r2_score(y_test, dl_preds)
dl_accuracy = max(0, dl_r2 * 100)

results.append({
    "Model": "Deep Learning",
    "MAE": dl_mae,
    "MSE": dl_mse,
    "RMSE": dl_rmse,
    "R2_Score": dl_r2,
    "Accuracy_Percentage": dl_accuracy
})

predictions_dict["Deep Learning"] = dl_preds

# =========================================================
# RESULT CSV
# =========================================================
results_df = pd.DataFrame(results)

result_csv_path = os.path.join(output_dir, "regional_inclusion_result.csv")
results_df.to_csv(result_csv_path, index=False)

print("\nModel Results")
print(results_df)

# =========================================================
# PREDICTION CSV
# =========================================================
prediction_df = pd.DataFrame({
    "District": district_test.values,
    "Actual_2011_Total_SC_Percentage": y_test.values
})

for model_name, preds in predictions_dict.items():
    prediction_df[f"{model_name}_Prediction"] = preds

prediction_csv_path = os.path.join(output_dir, "regional_inclusion_prediction.csv")
prediction_df.to_csv(prediction_csv_path, index=False)

print("\nPrediction Results")
print(prediction_df.head())

# =========================================================
# SAVE H5 MODEL
# =========================================================
h5_path = os.path.join(output_dir, "regional_inclusion_model.h5")
dl_model.save(h5_path)

# =========================================================
# SAVE PKL MODEL
# =========================================================
best_model_name = results_df.sort_values(by="R2_Score", ascending=False).iloc[0]["Model"]

if best_model_name == "Deep Learning":
    best_model = dl_model
else:
    best_model = models[best_model_name]

pkl_data = {
    "best_model_name": best_model_name,
    "models": models,
    "scaler": scaler,
    "district_encoder": district_encoder,
    "feature_columns": feature_cols,
    "target_column": target_col
}

pkl_path = os.path.join(output_dir, "regional_inclusion_model.pkl")

with open(pkl_path, "wb") as f:
    pickle.dump(pkl_data, f)

# =========================================================
# SAVE JSON
# =========================================================
json_path = os.path.join(output_dir, "regional_inclusion_results.json")

json_data = {
    "project_name": "Smart Regional Inclusion Analysis System",
    "author": "Sagnik Patra",
    "dataset_path": csv_path,
    "output_directory": output_dir,
    "target_column": target_col,
    "feature_columns": feature_cols,
    "best_model": best_model_name,
    "model_results": results_df.to_dict(orient="records"),
    "files_saved": {
        "h5_model": h5_path,
        "pkl_model": pkl_path,
        "result_csv": result_csv_path,
        "prediction_csv": prediction_csv_path
    }
}

with open(json_path, "w") as f:
    json.dump(json_data, f, indent=4)

# =========================================================
# SAVE YAML
# =========================================================
yaml_path = os.path.join(output_dir, "regional_inclusion_config.yaml")

yaml_data = {
    "project": "Smart Regional Inclusion Analysis System",
    "author": "Sagnik Patra",
    "model_type": "Regression",
    "target": target_col,
    "best_model": str(best_model_name),
    "features": feature_cols,
    "outputs": {
        "h5": h5_path,
        "pkl": pkl_path,
        "json": json_path,
        "yaml": yaml_path,
        "result_csv": result_csv_path,
        "prediction_csv": prediction_csv_path
    }
}

with open(yaml_path, "w") as f:
    yaml.dump(yaml_data, f, default_flow_style=False)

# =========================================================
# GRAPH 1: ACCURACY GRAPH
# =========================================================
plt.figure(figsize=(10, 6))
plt.bar(results_df["Model"], results_df["Accuracy_Percentage"])
plt.title("Model Accuracy Comparison")
plt.xlabel("Models")
plt.ylabel("Accuracy Percentage")
plt.xticks(rotation=30)
plt.tight_layout()

accuracy_graph_path = os.path.join(output_dir, "regional_inclusion_accuracy_graph.png")
plt.savefig(accuracy_graph_path, dpi=300)
plt.show()

# =========================================================
# GRAPH 2: HEATMAP
# =========================================================
corr = df[feature_cols + [target_col]].corr()

plt.figure(figsize=(12, 8))
plt.imshow(corr, cmap="coolwarm", aspect="auto")
plt.colorbar()
plt.xticks(range(len(corr.columns)), corr.columns, rotation=90)
plt.yticks(range(len(corr.columns)), corr.columns)
plt.title("Feature Correlation Heatmap")
plt.tight_layout()

heatmap_path = os.path.join(output_dir, "regional_inclusion_heatmap.png")
plt.savefig(heatmap_path, dpi=300)
plt.show()

# =========================================================
# GRAPH 3: COMPARISON GRAPH
# =========================================================
plt.figure(figsize=(10, 6))
plt.plot(results_df["Model"], results_df["MAE"], marker="o", label="MAE")
plt.plot(results_df["Model"], results_df["RMSE"], marker="o", label="RMSE")
plt.title("Model Error Comparison")
plt.xlabel("Models")
plt.ylabel("Error Value")
plt.xticks(rotation=30)
plt.legend()
plt.tight_layout()

comparison_graph_path = os.path.join(output_dir, "regional_inclusion_comparison_graph.png")
plt.savefig(comparison_graph_path, dpi=300)
plt.show()

# =========================================================
# GRAPH 4: RESULT GRAPH
# =========================================================
plt.figure(figsize=(10, 6))
plt.bar(results_df["Model"], results_df["R2_Score"])
plt.title("R2 Score Result Graph")
plt.xlabel("Models")
plt.ylabel("R2 Score")
plt.xticks(rotation=30)
plt.tight_layout()

result_graph_path = os.path.join(output_dir, "regional_inclusion_result_graph.png")
plt.savefig(result_graph_path, dpi=300)
plt.show()

# =========================================================
# GRAPH 5: PREDICTION GRAPH
# =========================================================
best_prediction_col = f"{best_model_name}_Prediction"

plt.figure(figsize=(12, 6))
plt.plot(prediction_df["Actual_2011_Total_SC_Percentage"].values, marker="o", label="Actual")
plt.plot(prediction_df[best_prediction_col].values, marker="o", label="Predicted")
plt.title(f"Actual vs Predicted SC Percentage using {best_model_name}")
plt.xlabel("Test Samples")
plt.ylabel("2011 Total SC Percentage")
plt.legend()
plt.tight_layout()

prediction_graph_path = os.path.join(output_dir, "regional_inclusion_prediction_graph.png")
plt.savefig(prediction_graph_path, dpi=300)
plt.show()

# =========================================================
# GRAPH 6: DISTRICT GROWTH GRAPH
# =========================================================
top_growth_df = df.sort_values(by="Total_Growth", ascending=False).head(15)

plt.figure(figsize=(12, 6))
plt.bar(top_growth_df[district_col], top_growth_df["Total_Growth"])
plt.title("Top 15 Districts by SC Population Growth")
plt.xlabel("District")
plt.ylabel("Total Growth")
plt.xticks(rotation=75)
plt.tight_layout()

growth_graph_path = os.path.join(output_dir, "regional_inclusion_growth_graph.png")
plt.savefig(growth_graph_path, dpi=300)
plt.show()

# =========================================================
# SAVE TRAINING HISTORY GRAPH
# =========================================================
plt.figure(figsize=(10, 6))
plt.plot(history.history["loss"], label="Training Loss")
plt.plot(history.history["val_loss"], label="Validation Loss")
plt.title("Deep Learning Training Loss Graph")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.legend()
plt.tight_layout()

training_graph_path = os.path.join(output_dir, "regional_inclusion_training_graph.png")
plt.savefig(training_graph_path, dpi=300)
plt.show()

# =========================================================
# FINAL OUTPUT
# =========================================================
print("\nAll files saved successfully")
print("H5 Model:", h5_path)
print("PKL Model:", pkl_path)
print("JSON File:", json_path)
print("YAML File:", yaml_path)
print("Result CSV:", result_csv_path)
print("Prediction CSV:", prediction_csv_path)
print("Accuracy Graph:", accuracy_graph_path)
print("Heatmap:", heatmap_path)
print("Comparison Graph:", comparison_graph_path)
print("Result Graph:", result_graph_path)
print("Prediction Graph:", prediction_graph_path)
print("Growth Graph:", growth_graph_path)
print("Training Graph:", training_graph_path)
```
